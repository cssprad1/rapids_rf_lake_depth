# Correlation Analysis for Lake Depth data

```python
from load_dataset import custom_lakedepth
import pandas as pd
import numpy as np
```

```python
primary_dataset = custom_lakedepth.LakeDepth(42)
pd_covariates = primary_dataset.covariates.to_pandas()
pd_covariates.describe()
```

## Just checking out our covariates

```python
pd_covariates.head(5)
```

## Make the correlation matrix

```python
corr = pd_covariates.corr()
corr.style.background_gradient(cmap='coolwarm', axis=None)
```

## Calculate correlation for depth and 1-6

```python
depth_one_six = primary_dataset.data_df.to_pandas()
depth_one_six.head(5)
```

```python
p_cols = depth_one_six.columns[:7]
depth_one_six_adjusted = depth_one_six[p_cols]
depth_one_six_adjusted.head(5)
```

## Correlation between depth and bands 1-6

```python
corr_one_six = depth_one_six_adjusted.corr()
corr_one_six.style.background_gradient(cmap='coolwarm', axis=None)
```
